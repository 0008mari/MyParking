# 주차어때 - 주차장정보 가공

```python
import pandas as pd
```

```python
data = pd.read_csv("서울시 공영주차장 안내 정보_20221101.csv", encoding='utf-8')
```

```python
columns = ['주차장 종류명', '운영구분명', '총 주차면', '유무료구분명', '평일 운영 시작시각(HHMM)', '평일 운영 종료시각(HHMM)', '기본 주차 요금', '기본 주차 시간(분 단위)', '추가 단위 요금', '추가 단위 시간(분 단위)', '전화번호', '주차장코드', '주차장명', '주소'] 
data2 = data[columns].copy()
data2.head(3)
```

```python
new_columns = ['parking_type_name', 'parking_operation_rule_name', 'parking_capacity', 'parking_pay_name', 'parking_weekday_begin_time', 'parking_weekday_end_time', 'parking_rates', 'parking_time_rates', 'parking_add_rates', 'parking_add_time_rates', 'parking_tel', 'parking_code', 'parking_name', 'parking_address']

data2.columns = new_columns
```

```python
# parking_code를 기준으로 중복 제거
data2 = data2.drop_duplicates(['parking_code'])
data2 = data2.reset_index(drop=True)
data2[50:55]
```

```python
# csv로 다시 저장

data2.to_csv('parkings_20221101.csv')
```
